```python
#file upload
from google.colab import files
uploaded = files.upload()
```

```python
#dataframe creation
import pandas as pd
import io
import seaborn as sns
import numpy as np
import datetime
df = pd.read_csv(io.BytesIO(uploaded['case study.csv']))
df.head()
```

```python
#info about the columns
df.info()
```

```python
#converting the datatypes of some columns
df['session_time'] = df['session_time'].apply(pd.to_datetime)
df['student_gender']=df['student_gender'].astype('category').cat.codes
df['family_status']=df['family_status'].astype('category').cat.codes
df['graduation_obstacle']=df['graduation_obstacle'].astype('category').cat.codes
df.info()
```

```python
#finding correlation between student_attended and other columns
corr_matrix = df.corr()['student_attended']
corr_matrix
```
